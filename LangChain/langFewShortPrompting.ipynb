```python

from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate, FewShotChatMessagePromptTemplate
#from langchain_community.chat_models.ollama import ChatOllama
```

```python
examples = [
    {"input": "India", "output": "aidnI"},
    {"input": "Germany", "output": "ynamreG"},
    {"input": "United States", "output": "setatS detinU"},
    {"input": "Canada", "output": "adanaC"}
]
```

```python
example_prompt = ChatPromptTemplate.from_messages([
    ("human", "Reverse the country name given to you in the examples below."),
    ("human", "{input}"),
    ("ai", "{output}")
])
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
)
```

```python
print(few_shot_prompt.format())
```

```python
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a linguistic specialist."),
    few_shot_prompt,
    ("human", "{input}")
])
```

```python
prompt_template
```

```python
prompt=prompt_template.format_messages(input="china")
```

```python
prompt
```

```python
llm = OllamaLLM(model="llama3.2", temperature=0.2)
response = llm.invoke(prompt)
```
