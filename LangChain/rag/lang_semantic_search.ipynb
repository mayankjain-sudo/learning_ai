```python
#%pip install langchain-chroma

from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma
```

```python
embeddings = OllamaEmbeddings(model="embeddinggemma")
```

```python
docs = [
    "Thrilling Finale awaits: The Coundown to the Cricket World Cup Championship begins!",
    "Global Giants Clash: Football World Cup Heats Up with Unforgettable Matches!",
    "Records Crowds and Unforgottable Moments: Highlights from the Cricket World Cup",
]
```

```python
vector_store = Chroma.from_texts(docs, embedding=embeddings)
```

```python
vector_store.similarity_search("Cricket", k=2)
```

```python
vector_store.similarity_search("football", k=2)
```
