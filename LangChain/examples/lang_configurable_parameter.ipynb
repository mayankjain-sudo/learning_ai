Exploring the configurable parameter in LCEL. Basically it send parameter at run time.

```python
from langchain_ollama import OllamaLLM, ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import ConfigurableField
```

```python
llm = ChatOllama(model="llama3.2").configurable_fields(
    model=ConfigurableField(
        description="The name of the LLM model to use",
        id="model",
        name="model"
    )
)
```

```python
prompt = PromptTemplate.from_template("Tell me a joke about {topic}.")
```

```python
chain = prompt | llm
chain.invoke({"topic": "cat"})
```

```python
chain.with_config(configurable={"model": "deepseek-r1:7b"}).invoke({"topic": "cat"})
```
