Langchain csv parser

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate
#from langchain.output_parsers import CommaSeparatedListOutputParser
from langchain_core.output_parsers.list import CommaSeparatedListOutputParser
#from langchain.output_parsers.list import ListOutputParser
from langchain_core.output_parsers.list import ListOutputParser
```

```python
llm  = OllamaLLM(model="llama3.2", temperature=0.6)
```

```python
prompt = PromptTemplate(
    input_variables=["things"],
    template="List 3 {things}"
)
llm.invoke(prompt.format(things="Countries that play cricket in world cup"))
```

```python
output_parser = CommaSeparatedListOutputParser()
formatted_output = output_parser.get_format_instructions()
```

```python
print(formatted_output)
```

```python
prompt = PromptTemplate(
    input_variables=["things"],
    template="List 3 {things}.\n {format_instructions}",
    partial_variables={"format_instructions": formatted_output}
)
```

```python
final_prompt = prompt.format(things="Countries that play cricket in world cup")
print(final_prompt)
```

```python
output = llm.invoke(input=final_prompt)
print(output.strip())
```

```python
type(output)
```

```python
things = output_parser.parse(output)
print(things)
```
