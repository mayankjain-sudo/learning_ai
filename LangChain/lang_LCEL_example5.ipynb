Multiple chain example

```python

from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
title = (
        ChatPromptTemplate.from_template(
            "Generate a impactful and engaging title for {input}" )
        | OllamaLLM(model="llama3.2")
        | StrOutputParser()
        | {"title": RunnablePassthrough()}
)
```

```python
outline = (
    ChatPromptTemplate.from_template(
        "Create a detailed outline for {title}"
    )
    | OllamaLLM(model="llama3.2")
    | StrOutputParser()
    | {"outline": RunnablePassthrough()}
)
```

```python
blog = (
    ChatPromptTemplate.from_template(
        "Write a comprehensive 200 words blog post based on the following outline: {outline}"
    )
    | OllamaLLM(model="llama3.2")
    | StrOutputParser()
    | {"blog_post": RunnablePassthrough()}
)
```

```python
summary = (
    ChatPromptTemplate.from_template(
        "Summarize the following blog post in a concise manner: {blog_post}"
    )
    | OllamaLLM(model="llama3.2")
    | StrOutputParser()
)
```

```python
content_chain = title | outline | blog | summary
```

```python
content_chain.invoke({"input": "The impact of AI on jobs"})
```
