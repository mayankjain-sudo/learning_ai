Bringing the history capability to LLM 

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
```

```python
llm = OllamaLLM(model="llama3.2")
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant who's good at {ability}. Respond in 20 words or fewer",
        ),
        ("human", "{input}"),
        MessagesPlaceholder(variable_name="history"),
    ]
)

base_chain = prompt | llm
```

```python
history = []
base_chain.invoke(
    { "ability": "math", "input": "What is a right angled triangle?", "history": [] }
)
```

```python
history.append(
    {
        "role": "human",
        "content": "What is a right angled triangle?",
    }
)
history.append(
    {
        "role": "assistant",
        "content": "A right angled triangle has one angle of 90 degrees. The side opposite this angle is the hypotenuse.",
    }
)
```

```python
base_chain.invoke(
    {
        "input": "what are the other types?",
        "ability": "math",
        "history": history
    }
)
```
