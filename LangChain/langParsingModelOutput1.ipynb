Langchain Json parser example

```python
# Ensure compatible versions of langchain packages are installed
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers.list import CommaSeparatedListOutputParser
from langchain_core.output_parsers.list import ListOutputParser
from langchain_core.output_parsers import JsonOutputParser
import json
```

```python
llm  = OllamaLLM(model="deepseek-r1:7b", temperature=0.3)
```

```python
prompt = PromptTemplate(
    input_variables=["continent"],
    template="List 3 countried in {continent} and their capitals."
)
print(llm.invoke(input=prompt.format(continent="Asia")))
```

```python
output_parser = JsonOutputParser()
formatted_output = output_parser.get_format_instructions()
print(formatted_output)
```

```python
prompt = PromptTemplate(
    input_variables=["continent"],
    template="List 3 countried in {continent} and their capitals\n{format_instructions}",
    partial_variables={"format_instructions": formatted_output}
)
prompt
```

```python
print(prompt.format(continent="North America"))
```

```python
output = llm.invoke(input=prompt.format(continent="North America"))
print(output)
type(output)
```

```python
countries=output_parser.parse(output)
print(countries)
```
